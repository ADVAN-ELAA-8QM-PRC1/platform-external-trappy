# Quick Start

To analyze a `trace.txt` that you have as the result of a run, you need to know the cpu mask of the clusters and the order of the actors as seen by the power allocator governor. This is platform specific information that can't be automatically guessed by `cr2`, so we need to tell it about them. For the cpumasks, look at the trace to know the masks, for example:
    
    $ grep -o '\Wcpus=[0-9a-f,]\+' trace.txt | sort -u
     cpus=0000000f
     cpus=000000f0
     
For this system we know that cpus 0-3 are the A7 cluster and cpus 4-7 are the A15 cluster. We call this the map_label and for this system that is:

```python
map_label = {"0000000f": "A7", "000000f0": "A15"}
```

We will need this later. We also need to know how many actors are in the system and their order. We can get the number of actors from the trace:

    $ grep -m 1 -o 'req_power={[0-9 ]\+}' trace.txt
    req_power={0 1581 4112}

This tells us that there are three actors. We need to guess which one they are and their order, you can usually do that by looking at the data. For this example, we'll assume the first one is the GPU and the others the A15 and A7 cluster:

```python
actor_order = ["GPU", "A15", "A7"]
```

To analyze a `trace.txt` that you may have stored in `/tmp` as the result of a run add the following to a cell to a new ipython notebook. Substitute `map_label` and `actor_order` with the ones you calculated before:

```python
%pylab inline
import sys
sys.path.append("..")
import cr2

map_label = {"0000000f": "A7", "000000f0": "A15"}
actor_order = ["GPU", "A15", "A7"]

cr2.summary_plots(actor_order, map_label, path="/tmp")
```

```python
import sys,os
import tempfile
import shutil
sys.path.append("..")
%pylab inline
import cr2
trace_thermal = "./trace.txt"
trace_sched = "../tests/trace_sched.txt"
TEMP_BASE = "/tmp"
```

```python

def setup_thermal():
    tDir = tempfile.mkdtemp(dir="/tmp", prefix="cr2_doc", suffix = ".tempDir")
    shutil.copyfile(trace_thermal, os.path.join(tDir, "trace.txt"))
    return tDir

def setup_sched():
    tDir = tempfile.mkdtemp(dir="/tmp", prefix="cr2_doc", suffix = ".tempDir")
    shutil.copyfile(trace_sched, os.path.join(tDir, "trace.txt"))
    return tDir
```

```python
temp_thermal_location = setup_thermal()
run1 = cr2.Run(temp_thermal_location)
run2 = cr2.Run(temp_thermal_location)
run2.thermal.data_frame["temp"] = run1.thermal.data_frame["temp"] * 2
run2.cpu_out_power.data_frame["power"] = run1.cpu_out_power.data_frame["power"] * 2
```

# No Pivot Simple

```python
        l = cr2.LinePlot(run1, cr2.thermal.Thermal, column="temp")
        l.view()
```

# No Pivot Multi Run

```python
        l = cr2.LinePlot(
            [run1, run2], cr2.thermal.Thermal, column="temp")
        l.view()
```

# No Pivot Mutiple Columns and Runs

```python
        l = cr2.LinePlot([run1,
                                  run2],
                                 [cr2.thermal.Thermal,
                                  cr2.thermal.ThermalGovernor],
                                 column=["temp",
                                         "power_range"])
        l.view()
```

# No Pivot With Filters

```python
        l = cr2.LinePlot([run1,
                                  run2],
                                 [cr2.cpu_power.CpuOutPower],
                                 column="power",
                                 filters={"cdev_state": [0]})
        l.view()
```

# Pivoted Data

```python
        l = cr2.LinePlot(
            run1,
            cr2.thermal.Thermal,
            column="temp",
            pivot="thermal_zone")
        l.view()
```

# Pivoted Multi Run Data

```python
        l = cr2.LinePlot(
            [run1, run2], cr2.cpu_power.CpuOutPower, column="power", pivot="cpus")
        l.view()
```

# Pivoted Data Multiple Runs and Columns

```python
        l = cr2.LinePlot([run1,
                                  run2],
                                 [cr2.cpu_power.CpuInPower,
                                  cr2.cpu_power.CpuOutPower],
                                 column=["dynamic_power", "power"],
                                 pivot="cpus")
        l.view()
```

# Pivoted Data with filters

```python
        l = cr2.LinePlot(
            run1,
            cr2.cpu_power.CpuInPower,
            column=[
                "dynamic_power",
                "load1"],
            filters={
                "cdev_state": [
                    1,
                    0]},
            pivot="cpus")
        l.view()
```

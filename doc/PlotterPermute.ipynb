```python
import sys,os
import tempfile
import shutil
sys.path.append("..")
%matplotlib inline
import trappy
trace_thermal = "./trace.txt"
trace_sched = "../tests/trace_sched.txt"
TEMP_BASE = "/tmp"
```

```python
def setup_thermal():
    tDir = tempfile.mkdtemp(dir="/tmp", prefix="trappy_doc", suffix = ".tempDir")
    shutil.copyfile(trace_thermal, os.path.join(tDir, "trace.txt"))
    return tDir

def setup_sched():
    tDir = tempfile.mkdtemp(dir="/tmp", prefix="trappy_doc", suffix = ".tempDir")
    shutil.copyfile(trace_sched, os.path.join(tDir, "trace.txt"))
    return tDir
```

```python
temp_thermal_location = setup_thermal()
trace1 = trappy.FTrace(temp_thermal_location, name="first trace")
```

# Permute Multi Trace with Multi Columns

The permute parameter allows to plot all the permutations of the specified traces with the specified columns. Each figure will plot the values of each column for those traces. Hence, with N traces and M columns, you will get M figures each displaying N traces.

Data in *trace1* looks like this:

```python
trace1.thermal.data_frame
```

For simplicity, let's build a second trace from the first one

```python
trace2 = trappy.FTrace(temp_thermal_location, name="second trace")
trace2.thermal.data_frame["temp"] = trace1.thermal.data_frame["temp"] * 2
trace2.thermal.data_frame["temp_prev"] = trace1.thermal.data_frame["temp_prev"] + 2000
trace2.cpu_in_power.data_frame["dynamic_power"] = trace1.cpu_in_power.data_frame["dynamic_power"] * 2
trace2.cpu_in_power.data_frame["static_power"] = trace1.cpu_in_power.data_frame["static_power"] * 2
```

```python
l = trappy.LinePlot([trace1, trace2], trappy.thermal.Thermal, column=["temp", "temp_prev"], permute=True)
l.view()
```

As you can see we obtained two figures, as we specified two columns:
- First figure shows the *temp* column for both traces
- Second figure shows the *temp_prev* column for both traces

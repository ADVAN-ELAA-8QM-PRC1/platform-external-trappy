# The Run class

The `Run` class allows you to inspect the data for the run.  Its main purpose is to parse the trace information into pandas DataFrames.  Each tracepoint gets its own DataFrame.

First some hacks to allow importing cr2 from here

```python
import sys
sys.path.append("..")
```

```python
%pylab inline
import cr2
```

The `Run` class has one mandatory parameter: the path to `trace.dat`

```python
run = cr2.Run("../tests")
```

You can access the DataFrame for each tracepoint using the `.data_frame` property.  For example, the `trace_thermal` tracepoint is in the `.thermal` property of `Run`

```python
run.thermal.data_frame
```

You can use pandas' infrastructure to filter data.  For example, the `in_power` trace has information for both clusters

```python
run.in_power.data_frame.head()
```

We can filter it to return only the data for cpus "000000f0":

```python
run.in_power.data_frame[run.in_power.data_frame.cpus == "000000f0"]
```

## Writing to CSV

You can use the `.write_csv()` method of each tracepoint class to output the trace to CSV for analyzing with other programs.  For example, let's write the `trace_thermal_power_allocator` tracepoint information (available in the `.thermal_governor` property of the `Run` instance) to file `thermal_power_allocator.csv`

```python
run.thermal_governor.write_csv("thermal_power_allocator.csv")
```

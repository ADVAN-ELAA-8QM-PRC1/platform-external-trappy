`map_label` is a dict that maps cpumasks to a name.  This name will be used as a label for the load and frequency plots.

```python
map_label = {"0000000f": "A7", "000000f0": "A15"}
```

`actor_order` is an array that with the order in which the actors appear in the `thermal_power_allocator` tracepoint.

```python
actor_order = ["GPU", "A15", "A7"]
```

`runs` is an array of tuples consisting of a name and the path to the directory where the `trace.dat` can be found.

```python
runs = [("good", "../tests"), ("experiment", "../tests")]
```

```python
cr2.compare_runs(actor_order, map_label, runs)
```
